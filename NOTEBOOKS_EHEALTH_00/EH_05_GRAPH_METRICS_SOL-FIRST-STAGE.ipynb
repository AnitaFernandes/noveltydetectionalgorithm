### COMPARANDO MELHOR DESEMPENHO DOS ALGORITMOS

```python
#ACURACIA
sns.set_theme(font_scale=1.3)
ax = sns.lineplot(x="semanas", y="acuracia", data=mlof_dfreq, color='blue')
ax = sns.lineplot(x="semanas", y="acuracia", data=miforest_dfreq, color='red')
ax = sns.lineplot(x="semanas", y="acuracia", data=mocsvm_dfreq, color='magenta')
ax = sns.lineplot(x="semanas", y="acuracia", data=mrce_dfreq, color='green')
ax.set(xlabel ="Semanas", ylabel = "Acurcia", title ='Acurcia dos modelos')
plt.gcf().subplots_adjust(bottom=0.15, left =0.15)
plt.savefig('models_acuracia.svg')
```

```python
#F1-SCORE
sns.set_theme(font_scale=1.3)
ax = sns.lineplot(x="semanas", y="precisao", data=mlof_dfreq, color='blue')
ax = sns.lineplot(x="semanas", y="precisao", data=miforest_dfreq, color='red')
ax = sns.lineplot(x="semanas", y="precisao", data=mocsvm_frequencia, color='magenta')
ax = sns.lineplot(x="semanas", y="precisao", data=mrce_dfreq, color='green')
ax.set(xlabel ="Semanas", ylabel = "Preciso", title ='Preciso dos modelos')
plt.gcf().subplots_adjust(bottom=0.15, left =0.15)
plt.savefig('models_precis.svg')
```

```python
#F1-SCORE
sns.set_theme(font_scale=1.3)
ax = sns.lineplot(x="semanas", y="sensibilidade", data=mlof_dfreq, color='blue')
ax = sns.lineplot(x="semanas", y="sensibilidade", data=miforest_dfreq, color='red')
ax = sns.lineplot(x="semanas", y="sensibilidade", data=mocsvm_frequencia, color='magenta')
ax = sns.lineplot(x="semanas", y="sensibilidade", data=mrce_dfreq, color='green')
ax.set(xlabel ="Semanas", ylabel = "Recall", title ='Recall dos modelos')
plt.gcf().subplots_adjust(bottom=0.15, left =0.15)
plt.savefig('models_recall.svg')
```

```python
#F1-SCORE
sns.set_theme(font_scale=1.3)
ax = sns.lineplot(x="semanas", y="f1score", data=mlof_dfreq, color='blue')
ax = sns.lineplot(x="semanas", y="f1score", data=miforest_dfreq, color='red')
ax = sns.lineplot(x="semanas", y="f1score", data=mocsvm_frequencia, color='magenta')
ax = sns.lineplot(x="semanas", y="f1score", data=mrce_dfreq, color='green')
ax.set(xlabel ="Semanas", ylabel = "F1-Score", title ='F1-Score dos modelos')
plt.gcf().subplots_adjust(bottom=0.15, left =0.15)
plt.savefig('models_f1.svg')
```

## Concatenando Dataframes

```python
dfmetricas = pd.concat([mlof_dfreq, miforest_dfreq,mocsvm_frequencia,mrce_dfreq],ignore_index=True)
print(dfmetricas.shape)
dfmetricas.head()
dfmetricas['tecnica'] = dfmetricas['algoritmo'] +'_'+ dfmetricas['variavel']
dfmetricas.head()
```

```python
ax = sns.barplot(x="semanas", y="acuracia", hue='tecnica', data=dfmetricas)
ax.set(xlabel ="Semanas", ylabel = "Acurcia", title ='Acurcia dos modelos')
ax.set(ylim=(0.5, 1.02))
ax.set(xlim=(-0.5, 8.5))
plt.gcf().subplots_adjust(bottom=0.15, left =0.15)
```

```python
ax = sns.barplot(x="semanas", y="acuracia", hue='tecnica', data=dfmetricas)
ax.set(xlabel ="Semanas", ylabel = "Acurcia", title ='Acurcia dos modelos')
ax.set(ylim=(0.5, 1.02))
ax.set(xlim=(-0.5, 8.5))
plt.gcf().subplots_adjust(bottom=0.15, left =0.15)
plt.legend([],[], frameon=False)
plt.savefig('bar_models_acuracia.svg')
```

```python
ax = sns.barplot(x="semanas", y="precisao", hue='tecnica', data=dfmetricas)
ax.set(xlabel ="Semanas", ylabel = "Preciso", title ='Preciso dos modelos')
ax.set(ylim=(0.5, 1.02))
ax.set(xlim=(-0.5, 8.5))
plt.gcf().subplots_adjust(bottom=0.15, left =0.15)
plt.legend([],[], frameon=False)
plt.savefig('bar_models_precisao.svg')
```

```python
ax = sns.barplot(x="semanas", y="sensibilidade", hue='tecnica', data=dfmetricas)
ax.set(xlabel ="Semanas", ylabel = "Recall", title ='Recall dos modelos')
ax.set(ylim=(0.5, 1.02))
ax.set(xlim=(-0.5, 8.5))
plt.gcf().subplots_adjust(bottom=0.15, left =0.15)
plt.legend([],[], frameon=False)
plt.savefig('bar_models_recall.svg')
```

```python
ax = sns.barplot(x="semanas", y="f1score", hue='tecnica', data=dfmetricas)
ax.set(xlabel ="Semanas", ylabel = "F1-Score", title ='F1-Score dos modelos')
ax.set(ylim=(0.5, 1.02))
ax.set(xlim=(-0.5, 8.5))
plt.gcf().subplots_adjust(bottom=0.15, left =0.15)
plt.legend([],[], frameon=False)
plt.savefig('bar_models_f1.svg')
```
