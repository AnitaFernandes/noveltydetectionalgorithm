## Treinando MODELOS (4) para SEMANAS (20) para DURACAO

```python
# experimento Classificação de Dias Anômalos

for z in range (1,4): #iterando atributos 1 duração, 2 frequência 
                      # e 3 duraçãp/frequência
    for y in range (1,5): #iteracao modelos 1 OC-SVM, 2 LOF, 3 iForest 
                          # e 4 RCE
        #inicializando listas temporárias para armazenamento das métricas
        csemana = []
        cvp = []
        cvn = []
        cfp = []
        cfn = []
        cacuracia = []
        cprecisao = []
        csensibilidade = []
        cf1escore = []
        for x in range(1,20): #iterando semanas 2 a 20 semanas (x+1) 
                              #de dados de treinamento
            semana = x+1
            if z == 1: #duracao
                p1testes = df_dur_test.iloc[:, [1,2,3,4,5,6,7,8]].copy()
                p1treins = df_dur_trein.iloc[0:semana*7].copy()
                p1treins = p1treins.iloc[:, [1,2,3,4,5,6,7,8]].copy()
                atributo = 'duracao'
            if z == 2: #frequencia
                p1testes = df_freq_test.iloc[:, [1,2,3,4,5,6,7,8]].copy()
                p1treins = df_freq_trein.iloc[0:semana*7].copy()
                p1treins = p1treins.iloc[:, [1,2,3,4,5,6,7,8]].copy()
                atributo = 'frequencia'
            if z == 3: #dfreq
                p1testes = df_dfreq_test.iloc[:, [1,2,3,4,5,6,7,8]].copy()
                p1treins = df_dfreq_trein.iloc[0:semana*7].copy()
                p1treins = p1treins.iloc[:, [1,2,3,4,5,6,7,8]].copy()
                atributo = 'dfreq'
            #standarização
            scaler = StandardScaler().fit(p1treins)
            p1treins = scaler.transform(p1treins).copy()
            p1testes = scaler.transform(p1testes).copy()

            # treinando modelo
            if y == 1:
                clf = svm.OneClassSVM(nu=0.1, kernel='rbf', gamma='scale')
                algoritmo = 'ocsvm'
            if y == 2:
                clf = LocalOutlierFactor(n_neighbors=10, novelty=True)
                algoritmo = 'lof'
            if y == 3:
                clf = IsolationForest(contamination= 0.01, random_state=1)
                algoritmo = 'iforest'
            if y == 4:
                clf = EllipticEnvelope(contamination= 0.01, random_state=1)
                algoritmo = 'rce'
            clf.fit(p1treins)
            #aplicando modelo
            y_pred_train = clf.predict(p1treins)
            y_pred_teste = clf.predict(p1testes)
            
            # calculando métricas
            n_error_train = y_pred_train[y_pred_train == -1].size
            n_error_teste = y_pred_teste[y_pred_teste ==  1].size  
            if z == 1 : #duracao
                df_dur_test['predicao'] = y_pred_teste
                df_dur_test['vp'] = np.where((df_dur_test['outlier'] == -1) & (df_dur_test['predicao'] == -1), 1,0)
                df_dur_test['vn'] = np.where((df_dur_test['outlier'] == 1) & (df_dur_test['predicao'] == 1), 1,0)
                df_dur_test['fp'] = np.where((df_dur_test['outlier'] == -1) & (df_dur_test['predicao'] == 1), 1,0)
                df_dur_test['fn'] = np.where((df_dur_test['outlier'] == 1) & (df_dur_test['predicao'] == -1), 1,0)
                VP = df_dur_test['vp'].sum()
                VN = df_dur_test['vn'].sum()
                FP = df_dur_test['fp'].sum()
                FN = df_dur_test['fn'].sum()
            if z == 2 : #freq
                df_freq_test['predicao'] = y_pred_teste
                df_freq_test['vp'] = np.where((df_freq_test['outlier'] == -1) & (df_freq_test['predicao'] == -1), 1,0)
                df_freq_test['vn'] = np.where((df_freq_test['outlier'] == 1) & (df_freq_test['predicao'] == 1), 1,0)
                df_freq_test['fp'] = np.where((df_freq_test['outlier'] == -1) & (df_freq_test['predicao'] == 1), 1,0)
                df_freq_test['fn'] = np.where((df_freq_test['outlier'] == 1) & (df_freq_test['predicao'] == -1), 1,0)
                VP = df_freq_test['vp'].sum()
                VN = df_freq_test['vn'].sum()
                FP = df_freq_test['fp'].sum()
                FN = df_freq_test['fn'].sum()                
            if z == 3 : #dfreq
                df_dfreq_test['predicao'] = y_pred_teste
                df_dfreq_test['vp'] = np.where((df_dfreq_test['outlier'] == -1) & (df_dfreq_test['predicao'] == -1), 1,0)
                df_dfreq_test['vn'] = np.where((df_dfreq_test['outlier'] == 1) & (df_dfreq_test['predicao'] == 1), 1,0)
                df_dfreq_test['fp'] = np.where((df_dfreq_test['outlier'] == -1) & (df_dfreq_test['predicao'] == 1), 1,0)
                df_dfreq_test['fn'] = np.where((df_dfreq_test['outlier'] == 1) & (df_dfreq_test['predicao'] == -1), 1,0)
                VP = df_dfreq_test['vp'].sum()
                VN = df_dfreq_test['vn'].sum()
                FP = df_dfreq_test['fp'].sum()
                FN = df_dfreq_test['fn'].sum()                
            erro = (FP+FN)/(VP+VN+FP+FN) #erro
            acuracia =  (VP+VN)/(VP+VN+FP+FN) # accuracy
            precisao = (VP)/(FP+VP) # precision
            sensibilidade = VP/(FN+VP) #recall
            f1escore = 2 * ((precisao * sensibilidade)/(precisao + sensibilidade))
             #armazenando metricas temp
            csemana.append(semana)
            cvp.append(VP)
            cvn.append(VN)
            cfp.append(FP)
            cfn.append(FN)
            cacuracia.append(acuracia)
            cprecisao.append(precisao)
            csensibilidade.append(sensibilidade)
            cf1escore.append(f1escore)
        #criando um dataframe a partir das métricas    
        zipped = list(zip(csemana,cvp,cvn,cfp,cfn,cacuracia, cprecisao, csensibilidade, cf1escore))    
        dfmetricas = pd.DataFrame(zipped, columns=['semanas', 'vp', 'vn', 'fp', 'fn','acuracia','precisao',
                                                   'sensibilidade','f1score'])
        dfmetricas['algoritmo'] = algoritmo
        dfmetricas['variavel'] = atributo
        dfmetricas['variacao'] = '20-60'
        print(dfmetricas.head(25))
        #exportando dataframes para datasets
        dfmetricas.to_csv(r'dataseteh/results/ehmetricas-'+algoritmo+'-'+atributo+'-trein01a20semanas-ativ8.csv', 
                          sep=';', index=False, decimal=',')

        #dentro do for Y MODELOS
    #dentro do for Z ATRIBUTOS
```
