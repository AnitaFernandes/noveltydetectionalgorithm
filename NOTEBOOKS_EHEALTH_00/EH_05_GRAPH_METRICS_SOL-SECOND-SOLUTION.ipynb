### Desempenho individual algoritmos 

```python
#***DESEMPENHO cclothes
```

```python
#ACURACIA
sns.set_theme(font_scale=1.3)
ax = sns.lineplot(x="semanas", y="acuracia", data=df_cclothes_lof, color='blue')
ax = sns.lineplot(x="semanas", y="acuracia", data=df_cclothes_if, color='red')
ax = sns.lineplot(x="semanas", y="acuracia", data=df_cclothes_ocsvm, color='magenta')
ax = sns.lineplot(x="semanas", y="acuracia", data=df_cclothes_rce, color='green')
ax.set(xlabel ="Semanas", ylabel = "Acurcia", title ='Atividade Change Clothes')
plt.gcf().subplots_adjust(bottom=0.15, left =0.15)
plt.savefig('ind_cclothes_acuracia.svg')
```

```python
#F1SCORE
sns.set_theme(font_scale=1.3)
ax = sns.lineplot(x="semanas", y="f1score", data=df_cclothes_lof, color='blue')
ax = sns.lineplot(x="semanas", y="f1score", data=df_cclothes_if, color='red')
ax = sns.lineplot(x="semanas", y="f1score", data=df_cclothes_ocsvm, color='magenta')
ax = sns.lineplot(x="semanas", y="f1score", data=df_cclothes_rce, color='green')
ax.set(xlabel ="Weeks", ylabel = "F1-Score", title ='Atividade Change Clothes')
plt.gcf().subplots_adjust(bottom=0.15, left =0.15)
plt.savefig('ind_cclothes_f1.svg')
```

```python
df_cclothes_metricas = pd.concat([df_cclothes_lof, df_cclothes_if,df_cclothes_ocsvm,df_cclothes_rce],ignore_index=True)
print(df_cclothes_metricas.shape)
df_cclothes_metricas.head()
```

```python
ax = sns.barplot(x="semanas", y="acuracia", hue='algoritmo', data=df_cclothes_metricas)
ax.set_xlabel("Semanas")
ax.set_ylabel("Acurcia")
```

```python
ax = sns.barplot(x="semanas", y="f1score", hue='algoritmo', data=df_cclothes_metricas)
ax.set_xlabel("Semanas")
ax.set_ylabel("F1-Score")
```
