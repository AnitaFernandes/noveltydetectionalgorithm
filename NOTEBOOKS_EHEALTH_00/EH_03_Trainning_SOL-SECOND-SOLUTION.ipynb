## Treinando MODELOS (4) para SEMANAS (20)

```python
for z in range (1,6): # iterando atividades 1 cclothes, 2 eating
                      # 3 mcoffe, 4 mhotfood, 5 toileting e 6 whandface
    for y in range (1,5): # iterando algoritmos 1 OC-SVM, 2 LOF, 3 iForest e 4 RCE
        # inicializando listas temporárias para armazenar as métricas
        csemana = []
        cvp = []
        cvn = []
        cfp = []
        cfn = []
        cacuracia = []
        cprecisao = []
        csensibilidade = []
        cf1escore = []
        for x in range(1,20): #iterando nas semanas 2 a 20 (x+1)
                              #de dados de treinamento
            semana = x+1
            if z==1: #cclothes
                p1testes = df_cclothes_test.iloc[:, [1,2,3,4,5]].copy()
                p1treins = df_cclothes_trein.iloc[0:semana*7].copy()
                p1treins = p1treins.iloc[:, [1,2,3,4,5]]
                atividade = 'cclothes'            
            if z==2: #eating
                p1testes = df_eating_test.iloc[:, [1,2,3,4,5]].copy()
                p1treins = df_eating_trein.iloc[0:semana*7].copy()
                p1treins = p1treins.iloc[:, [1,2,3,4,5]]
                atividade = 'eating'
            if z==3: #mcoffe
                p1testes = df_mcoffe_test.iloc[:, [1,2,3,4,5]].copy()
                p1treins = df_mcoffe_trein.iloc[0:semana*7].copy()
                p1treins = p1treins.iloc[:, [1,2,3,4,5]]
                atividade = 'mcoffe'
            if z==4: #mhotfood
                p1testes = df_mhotfood_test.iloc[:, [1,2,3,4,5]].copy()
                p1treins = df_mhotfood_trein.iloc[0:semana*7].copy()
                p1treins = p1treins.iloc[:, [1,2,3,4,5]]
                atividade = 'mhotfood'
            if z==5: #toileting
                p1testes = df_toileting_test.iloc[:, [1,2,3,4,5]].copy()
                p1treins = df_toileting_trein.iloc[0:semana*7].copy()
                p1treins = p1treins.iloc[:, [1,2,3,4,5]]                
                atividade = 'toileting'
            if z==6: #whandface
                p1testes = df_whandface_test.iloc[:, [1,2,3,4,5]].copy()
                p1treins = df_whandface_trein.iloc[0:semana*7].copy()
                p1treins = p1treins.iloc[:, [1,2,3,4,5]]                
                atividade = 'whandface'
            #standarização
            scaler = StandardScaler().fit(p1treins)
            p1treins = scaler.transform(p1treins).copy()
            p1testes = scaler.transform(p1testes).copy()

            # treinamento dos modelos
            if y == 1:
                clf = svm.OneClassSVM(nu=0.1, kernel='rbf', gamma='scale')
                algoritmo = 'ocsvm'
            if y == 2:
                clf = LocalOutlierFactor(n_neighbors=10, novelty=True)
                algoritmo = 'lof'
            if y == 3:
                clf = IsolationForest(contamination= 0.01, random_state=1)
                algoritmo = 'iforest'
            if y == 4:
                clf = EllipticEnvelope(contamination= 0.01, random_state=0, support_fraction= 0.9)
                algoritmo = 'rce'
            clf.fit(p1treins)
            y_pred_train = clf.predict(p1treins)
            y_pred_teste = clf.predict(p1testes)

            #extração das métricas
            n_error_train = y_pred_train[y_pred_train == -1].size
            n_error_teste = y_pred_teste[y_pred_teste ==  1].size
            # calculando métricas
            if z==1: #cclothes
                df_cclothes_test['predicao'] = y_pred_teste
                df_cclothes_test['vp'] = np.where((df_cclothes_test['outlier'] == -1) & (df_cclothes_test['predicao'] == -1), 1,0)
                df_cclothes_test['vn'] = np.where((df_cclothes_test['outlier'] == 1) & (df_cclothes_test['predicao'] == 1), 1,0)
                df_cclothes_test['fp'] = np.where((df_cclothes_test['outlier'] == -1) & (df_cclothes_test['predicao'] == 1), 1,0)
                df_cclothes_test['fn'] = np.where((df_cclothes_test['outlier'] == 1) & (df_cclothes_test['predicao'] == -1), 1,0)
                VP = df_cclothes_test['vp'].sum()
                VN = df_cclothes_test['vn'].sum()
                FP = df_cclothes_test['fp'].sum()
                FN = df_cclothes_test['fn'].sum()              
            if z==2: #eating               
                df_eating_test['predicao'] = y_pred_teste
                df_eating_test['vp'] = np.where((df_eating_test['outlier'] == -1) & (df_eating_test['predicao'] == -1), 1,0)
                df_eating_test['vn'] = np.where((df_eating_test['outlier'] == 1) & (df_eating_test['predicao'] == 1), 1,0)
                df_eating_test['fp'] = np.where((df_eating_test['outlier'] == -1) & (df_eating_test['predicao'] == 1), 1,0)
                df_eating_test['fn'] = np.where((df_eating_test['outlier'] == 1) & (df_eating_test['predicao'] == -1), 1,0)
                VP = df_eating_test['vp'].sum()
                VN = df_eating_test['vn'].sum()
                FP = df_eating_test['fp'].sum()
                FN = df_eating_test['fn'].sum()                
            if z==3: #mcoffe
                df_mcoffe_test['predicao'] = y_pred_teste
                df_mcoffe_test['vp'] = np.where((df_mcoffe_test['outlier'] == -1) & (df_mcoffe_test['predicao'] == -1), 1,0)
                df_mcoffe_test['vn'] = np.where((df_mcoffe_test['outlier'] == 1) & (df_mcoffe_test['predicao'] == 1), 1,0)
                df_mcoffe_test['fp'] = np.where((df_mcoffe_test['outlier'] == -1) & (df_mcoffe_test['predicao'] == 1), 1,0)
                df_mcoffe_test['fn'] = np.where((df_mcoffe_test['outlier'] == 1) & (df_mcoffe_test['predicao'] == -1), 1,0)
                VP = df_mcoffe_test['vp'].sum()
                VN = df_mcoffe_test['vn'].sum()
                FP = df_mcoffe_test['fp'].sum()
                FN = df_mcoffe_test['fn'].sum()                   
            if z==4: #mhotfood           
                df_mhotfood_test['predicao'] = y_pred_teste
                df_mhotfood_test['vp'] = np.where((df_mhotfood_test['outlier'] == -1) & (df_mhotfood_test['predicao'] == -1), 1,0)
                df_mhotfood_test['vn'] = np.where((df_mhotfood_test['outlier'] == 1) & (df_mhotfood_test['predicao'] == 1), 1,0)
                df_mhotfood_test['fp'] = np.where((df_mhotfood_test['outlier'] == -1) & (df_mhotfood_test['predicao'] == 1), 1,0)
                df_mhotfood_test['fn'] = np.where((df_mhotfood_test['outlier'] == 1) & (df_mhotfood_test['predicao'] == -1), 1,0)
                VP = df_mhotfood_test['vp'].sum()
                VN = df_mhotfood_test['vn'].sum()
                FP = df_mhotfood_test['fp'].sum()                
            if z==5: #toileting
                df_toileting_test['predicao'] = y_pred_teste
                df_toileting_test['vp'] = np.where((df_toileting_test['outlier'] == -1) & (df_toileting_test['predicao'] == -1), 1,0)
                df_toileting_test['vn'] = np.where((df_toileting_test['outlier'] == 1) & (df_toileting_test['predicao'] == 1), 1,0)
                df_toileting_test['fp'] = np.where((df_toileting_test['outlier'] == -1) & (df_toileting_test['predicao'] == 1), 1,0)
                df_toileting_test['fn'] = np.where((df_toileting_test['outlier'] == 1) & (df_toileting_test['predicao'] == -1), 1,0)
                VP = df_toileting_test['vp'].sum()
                VN = df_toileting_test['vn'].sum()
                FP = df_toileting_test['fp'].sum()                
            if z==6: #whandface
                df_whandface_test['predicao'] = y_pred_teste
                df_whandface_test['vp'] = np.where((df_whandface_test['outlier'] == -1) & (df_whandface_test['predicao'] == -1), 1,0)
                df_whandface_test['vn'] = np.where((df_whandface_test['outlier'] == 1) & (df_whandface_test['predicao'] == 1), 1,0)
                df_whandface_test['fp'] = np.where((df_whandface_test['outlier'] == -1) & (df_whandface_test['predicao'] == 1), 1,0)
                df_whandface_test['fn'] = np.where((df_whandface_test['outlier'] == 1) & (df_whandface_test['predicao'] == -1), 1,0)
                VP = df_whandface_test['vp'].sum()
                VN = df_whandface_test['vn'].sum()
                FP = df_whandface_test['fp'].sum()                   
            erro = (FP+FN)/(VP+VN+FP+FN) #erro
            acuracia =  (VP+VN)/(VP+VN+FP+FN) # accuracy
            precisao = (VP)/(FP+VP) # precision
            sensibilidade = VP/(FN+VP) #recall
            f1escore = 2 * ((precisao * sensibilidade)/(precisao + sensibilidade))
            #armazenando metricas temp
            csemana.append(semana)
            cvp.append(VP)
            cvn.append(VN)
            cfp.append(FP)
            cfn.append(FN)
            cacuracia.append(acuracia)
            cprecisao.append(precisao)
            csensibilidade.append(sensibilidade)
            cf1escore.append(f1escore)

        #criando um dataframe a partir das métricas    
        zipped = list(zip(csemana,cvp,cvn,cfp,cfn,cacuracia, cprecisao, csensibilidade, cf1escore))    
        dfmetricas = pd.DataFrame(zipped, columns=['semanas', 'vp', 'vn', 'fp', 'fn','acuracia',
                                                   'precisao','sensibilidade','f1score'])
        dfmetricas['algoritmo'] = algoritmo
        dfmetricas['variacao'] = '20-60'
        dfmetricas['atividade'] = atividade
        print(dfmetricas.head(20))
        #exportando dataframes para datasets
        dfmetricas.to_csv(r'dataseteh/results/ehmetricas-'+atividade+'-'+algoritmo+'.csv', sep=';', 
                          index=False, decimal=',')
```
