# Importing and pre-processing the outputP1.csv dataset


```python
## importando dataset outputPx.csv
df = pd.read_csv('dataseteh/pp1/outputP1.csv',delimiter=';',decimal=',',thousands=".",encoding='ANSI')
df.shape
```

```python
df.head()
```

```python
df['duracao_segundos'] = ((df['end_time_seconds'] - df['begin_time_seconds']))
df.head(20)
```

```python
dfnew = pd.DataFrame()
```

```python
dfnew['duracao'] = df.groupby(['day_begin','activity_number','activity_name'])['duracao_segundos'].sum()
dfnew.head()
```

```python
dfnew['frequencia'] = df.groupby(['day_begin','activity_number','activity_name'])['activity_number'].count()
dfnew.head()
```

```python
dfnew['durfreq'] = dfnew['duracao'] / dfnew['frequencia']
dfnew.head()
```

```python
dfnew = dfnew.reset_index()
```

```python
dfnew = dfnew.rename(columns={df.columns[0]: 'dia'})
```

```python
dfnew.insert(loc = 1, column = 'dia_semana', value = 0)
dfnew.insert(loc = 2, column = 'fim_semana', value = 0)
dfnew.head(20)
```

```python
# numerando dias da semana 0 a 6 sefunda a domingo - fim de semana 1 / dia de semana 0

dia = 0
cont = -1

for ind in dfnew.index:
    if (dfnew['dia'][ind] != dia) : #NOVO DIA
        dia = dfnew['dia'][ind]
        if cont == 6:
            cont = 0
        else:
            cont = cont+1
        dfnew['dia_semana'][ind] = cont
        if cont < 5:
            dfnew['fim_semana'][ind] = 0
        else:
            dfnew['fim_semana'][ind] = 1
    else: 
        dfnew['dia_semana'][ind] = cont
        if cont < 5:
            dfnew['fim_semana'][ind] = 0
        else:
            dfnew['fim_semana'][ind] = 1
```

```python
df = dfnew.copy()
df.head()
```

```python
df = dfpiv_freq.iloc[:,np.r_[0,1:3,4,5,9,10,18,22]]
df.head()
```
